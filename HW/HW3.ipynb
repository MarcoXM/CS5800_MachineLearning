```python
import numpy as np

def sig(x):
    return 1/(1 + np.exp(-x))
```

```python
def feedforward(x,layer1W,layer2W,layer3W):
    import numpy as np
    layer3W = layer3W.reshape(1,3)
    W1,b1 = layer1W[:,:layer1W.shape[1]-1], layer1W[:,-1] 
    W2,b2 = layer2W[:,:layer2W.shape[1]-1], layer2W[:,-1] 
    W3,b3 = layer3W[:,:layer3W.shape[1]-1], layer3W[:,-1] 
    #print(W3,b3)
    l1 = sig(W1.dot(x)+b1)
    #print(l1)

    l2 = sig(W2.dot(l1) + b2)
    #print(l2)
    out = sig(W3.dot(l2) + b3)
    return out

    
```

```python
def feedforwardRecurrent(x,layer1W,layer2W,layer3W):
    layer3W = layer3W.reshape(1,4)
    W1,w1p,b1 = layer1W[:,:layer1W.shape[1]-2],layer1W[:,-2],layer1W[:,-1] 
    W2,w2p,b2 = layer2W[:,:layer2W.shape[1]-2],layer2W[:,-2],layer2W[:,-1] 
    W3,w3p,b3 = layer3W[:,:layer3W.shape[1]-2],layer3W[:,-2],layer3W[:,-1]
    outpast = 0
    l1past = np.array([0,0]).reshape(2,)
    l2past = np.array([0,0]).reshape(2,)
    ans = []
    for t in range(x.shape[0]):
        l1 = sig(W1.dot(x[t]) + w1p*l1past +b1)
        l2 = sig(W2.dot(l1)+ w2p*l2past + b2)
        out = sig(W3.dot(l2) + outpast* w3p + b3)
        outpast = out.copy()
        l1past = l1.copy()
        l2past = l2.copy()
        ans.append(out)
    return ans
```

```python
def findWeights4(x,u):
    x = x.reshape(1,784)
    z = np.zeros(40).reshape(1,40)
    list_z = list(range(40))
    for i in range(40):
        index = 100
        tep = -10000000
        for j in list_z:
            pro = x.dot(u[:,j])
            if pro > tep:
                tep = pro
                index = j
        if tep < 0:
            break
        z[0][index] = tep
        list_z.remove(index)
        dx = (u[:,index] * tep).reshape(1,784)
        x -= dx
    
    return z
```
